# Viewing Data
Here is how to view the top and bottom rows of the frame:



```python
import numpy as np
import pandas as pd
```

```python
dates = pd.date_range('20130101', periods=10)
```

```python
df = pd.DataFrame(np.random.randn(10, 4), index=dates, columns=list('ABCD'))
```

```python
df
```

## Top and Bottom

```python
df.head()         #top data
```

```python
df.tail(3)      #bottom data
```

```python
df.index          #Display the index, columns:
```

```python
df.columns
```

# Some Other Operations

```python
df.describe()          #a quick statistic summary of data
```

```python
 df.T                   # transpose your data
```

```python
df.sort_index(axis=1, ascending=False)             # Sorting by an axis:
```

```python
df.sort_values(by='C')    # Sorting by values:
```

```python
df['B']                  # selecting sigle column
```

```python
df[2:4]              #Selecting via [], which slices the rows.
```

```python
df.iloc[1]                   #Select via the position of the passed integers:
```

```python
df.iloc[3:8, 0:2]           #By integer slices, acting similar to numpy/python:
```
